```python
## Finding out the number of toss wins the numbers of tose
ipl['toss_winner'].value_counts()
```

```python
## Extracting the record where a team won batting first
batting_fisrt=ipl[ipl['win_by_runs']!=0]
```

```python
## Looking at the head
batting_fisrt.head()
```

```python
## Making a histogram 
plt.figure(figsize=(7,7))
plt.hist(batting_fisrt['win_by_runs'])
plt.title("Distribution of Runs")
plt.xlabel('Runs')
plt.show()
```

```python
## Finding out the number of wins w.r.t each team after batting first
batting_fisrt['winner'].value_counts()
```

```python
#Making a bar-plot for top 3 teams with most wins after batting first
plt.figure(figsize=(7,7))
plt.bar(list(batting_fisrt['winner'].value_counts()[0:3].keys()),list(batting_fisrt['winner'].value_counts()[0:3]),color=['blue','yellow','orange'])
plt.xlabel("Top 3 Winner")
plt.show()
```

```python
#Making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_fisrt['winner'].value_counts()),labels=list(batting_fisrt['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
